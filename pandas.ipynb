```python
#NULL VALUES
#IN PANDAS MISSING DATA IS REPRESENTED BY TWO VALUES:
#NONE: IT IS A PYTHON SINGLETON OBJECT THAT IS OFTEN USED FOR MISSING DATA IN PYTHON CODE.
#NAN : IS A SPECIAL FLOATING - POINT VALUE RECOGNISED BY ALL SYSTEM THAT USE THE STANDARD IEEE FLOATING - POINT REPRESENTATIO
```

```python
import numpy as np
import pandas as pd
```

```python
#dictionary of Lists
dict = {'First Score':[100,90,np.nan,95],
     'Second Score':[30,45,32,np.nan],
     'Third Score':[np.nan,40,89,23]}

#CREATE A DATAFRAME FROM LIST
df = pd.DataFrame(dict)
df.isnull()
```

```python
#filling the missing values
```

```python
df.fillna(87)
```

```python
#will replace the NaN value in dataframe with the value - 99
df.replace(to_replace = np.nan, value = -99)
```

```python
#using dropna() function
df.dropna()
```
